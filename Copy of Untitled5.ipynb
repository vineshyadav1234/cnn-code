```python
df['director']=df['director'].fillna('director not specified')
df.head()
```

```python
df['director']=df['director'].replace('director not specified')
df.head()
```

```python
df=df.replace('derector not specified','director not specified')
df
```

```python
director_list=pd.DataFrame()
print(director_list)
```

```python
director_list=df['director'].str.split(',',expand=True).stack()
print(director_list)
```

```python
director_list=director_list.to_frame()# it is used to send the data in list
print(director_list)
```

```python
director_list.columns=['directors'] # it is used to  rename th ecol.
```

```python
director_list
```

```python
director=director_list.groupby(['directors']).size().reset_index(name='Total count')# this formula is used to count the value in columns
print(director)
```

```python
directors=director[director.directors !="director not specified"]
directors
```

```python
directors=directors.sort_values(by=['Total count'],ascending=False)
directors
```

```python
top5derectors=directors.head()
```

```python
top5derectors
```

```python
barchart=px.bar(top5derectors,x='Total count',y='directors',title='top 5 director of netflix')
barchart.show()
```
