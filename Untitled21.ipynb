```python
##any==how
df_pyspark.na.drop(how='any').show()
```

```python
##any==all
df_pyspark.na.drop(how='all').show()
```

```python
### tresold
df_pyspark.na.drop(how='any',thresh=2).show()## this is used if we set tresh is 2 then it will keep all that row which have non  2 null value
```

```python
## subset this is uesd when we want to delete a null value from any spesific col
df_pyspark.na.drop(how='any',subset=['experience']).show()
```

```python
### fill the missing value
df_pyspark.na.fill('missing value').show()
```

```python
### fill the missing value in specific col
df_pyspark.na.fill(0,['experience','age']).show()
```

```python
from pyspark.ml.feature import Imputer
imputer=Imputer(
    inputCols=['age','experience','salary'],
    outputCols=["{}_imputed".format(c) for c in ['age','experience','salary']]
).setStrategy("mean")
imputer.fit(df_pyspark).transform(df_pyspark).show()
df_pyspark.show()
```

```python
from pyspark.ml.feature import VectorAssembler
featureassembler=VectorAssembler(inputCols=['age','experience'],outputCol='independent feature')
```

```python
output=featureassembler.transform(df)
```

```python
output.show()
```

```python
finalized_data=output.select("independent feature",'salary')
```

```python
finalized_data.show()
```

```python
from pyspark.ml.regression import LinearRegression
train_data,test_data=finalized_data.randomSplit([0.75,0.25])
Regressor=LinearRegression(featuresCol='independent feature',labelCol='salary')
Regressor=Regressor.fit(train_data)
```

```python
Regressor.coefficients
```

```python
Regressor.intercept
```

```python
pre_result=Regressor.evaluate(test_data)
```

```python
pre_result.predictions.show()
```
